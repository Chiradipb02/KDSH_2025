```python
import os
from sentence_transformers import SentenceTransformer
from langchain.document_loaders import PyPDFLoader
from langchain.text_splitter import CharacterTextSplitter
from langchain.vectorstores import Chroma

# Initialize the SentenceTransformer model
model = SentenceTransformer('all-MiniLM-L6-v2')  # You can choose a different model from Hugging Face

# Directory containing PDF files
pdf_dir = "/content/drive/MyDrive/kgp/kdsh_data/Reference/Non-Publishable"

# Load and preprocess the documents
all_texts = []

# Loop through all PDF files in the directory
for pdf_file in os.listdir(pdf_dir):
    if pdf_file.endswith(".pdf"):
        file_path = os.path.join(pdf_dir, pdf_file)

        # Load PDF and extract text
        loader = PyPDFLoader(file_path)
        raw_documents = loader.load()

        # Split text into chunks
        text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=100)
        documents = text_splitter.split_documents(raw_documents)

        # Collect plain text chunks
        all_texts.extend([doc.page_content for doc in documents])

# Generate embeddings using SentenceTransformer
embeddings = model.encode(all_texts, convert_to_tensor=True)


```

```python
embeddings.shape
```

```python
from langchain.schema import Document

# Combine texts with their metadata
documents = [
    Document(page_content=text, metadata={"source": f"chunk_{i}"})
    for i, text in enumerate(all_texts)
]
db = Chroma.from_documents(documents, embeddings)

db.persist()
```

```python
from sentence_transformers import SentenceTransformer
from langchain.schema import Document
from langchain.vectorstores import Chroma
from langchain_community.embeddings import HuggingFaceEmbeddings



# Assuming all_texts is a list of strings
documents = [Document(page_content=text, metadata={"source": f"chunk_{i}"}) for i, text in enumerate(all_texts)]

# Generate embeddings using Sentence Transformers
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

# Create Chroma vectorstore
db = Chroma.from_documents(documents, embeddings)
db.persist()
```

```python
query = '''give the summary of the paper titled Transdimensional Properties of Graphite in Relation
to Cheese Consumption on Tuesday Afternoons ?
'''
docs = db.similarity_search(query)
print(docs[0].page_content)
```
