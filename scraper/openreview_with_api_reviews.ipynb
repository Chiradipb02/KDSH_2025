```python
import openreview
import os

# load OPENREVIEW_PASSWORD from environment variable
openreview_password = os.environ.get('OPENREVIEW_PASSWORD')
# load OFFICIAL_EMAIL from environment variable
official_email = os.environ.get('OFFICIAL_EMAIL')
```

```python
# Replace with your actual email and password
client = openreview.api.OpenReviewClient(
    baseurl='https://api2.openreview.net',
    username=official_email,
    password=openreview_password
)
```

```python
venue_id = 'NeurIPS.cc/2024/Conference'
venue_group = client.get_group(venue_id)
```

```python
for i in venue_group.content:
    print(i, venue_group.content[i])
```

```python
submissions = client.get_all_notes(invitation=f'{venue_id}/-/Submission', details='replies')
```

```python
len(submissions)
```

```python
print(submissions[0])
```

```python
from tqdm import tqdm
reviews ={}
for sub in tqdm(submissions):
    try:
        id = sub.id
        review = []
        for i in sub.details["replies"]:
            for j in i["content"]:
                if "weak" in j.lower():
                    review.append(i["content"][j]["value"])
        if len(review) > 0:
            reviews[id] = review
    except:
        pass
```

```python
len(reviews)
```

```python
# save reviews to a file
import json
with open('neurips_reviews.json', 'w') as f:
    json.dump(reviews, f, indent=4)
```
